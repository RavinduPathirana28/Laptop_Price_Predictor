```python
#changing the data types of Ram and Weigth to continues values 
data['Ram']=data['Ram'].str.replace('GB','').astype('int64')
data['Weight']=data['Weight'].str.replace('kg','').astype('float64')
```

```python
data['Company'].value_counts()
```

```python
def add_company(inpt):
    if inpt=='Samsung' or inpt=='Razer' or inpt=='Mediacom' or inpt=='Microsoft' or inpt=='Xiaomi' or inpt=='Vero' or inpt=='Chuwi' or inpt=='Google' or inpt=='Fujitsu' or inpt=='LG' or inpt=='Huawei':
        return 'Other'
    else:
        return inpt
data['Company']=data['Company'].apply(add_company)
```

```python
data['Company'].value_counts()
```

```python
data['Product'].value_counts()
```

```python
data['TypeName'].value_counts()
```

```python
data['ScreenResolution'].value_counts()
```

```python
data['Touchscreen']=data['ScreenResolution'].apply(lambda x:1 if 'Touchscreen' in x else 0)
data['IPS']=data['ScreenResolution'].apply(lambda x:1 if 'IPS' in x else 0)
```

```python
data.head()
```

```python
data['Cpu'].value_counts()
```

```python
data['cpu_name']=data['Cpu'].apply(lambda x: "".join(x.split()[0:3]))
data['cpu_name'].value_counts()
```

```python
def set_processor(c_name):
    if c_name=='IntelCorei7' or c_name=='IntelCorei5'or c_name=='IntelCorei3':
        return c_name
    else:
        if c_name.split()[0]=='AMD':
            return 'AMD'
        else:
            return 'Other'
data['cpu_name']=data['cpu_name'].apply(set_processor)
```

```python
data['cpu_name'].value_counts()
```

```python
data['Gpu'].value_counts()
```

```python
data['gpu_name']=data['Gpu'].apply(lambda x:"".join(x.split()[0:1]))
data['gpu_name'].value_counts()
```

```python
data.shape
```

```python
data=data[data['gpu_name'] !='ARM']
data.shape
```

```python

data.head()
```

```python
data['OpSys'].value_counts()
```

```python
def set_os(os):
    if os=='Windows 10' or os=='Windows 7' or os=='Windows 10 S':
        return 'Windows'
    elif os=='Linux':
        return os
    elif os=='macOS' or os=='Mac OS X':
        return 'Mac'
    else:
        return 'Other'
data['OpSys']=data['OpSys'].apply(set_os)
data['OpSys'].value_counts()
```

```python
data=data.drop(columns=['laptop_ID','ScreenResolution','Cpu','Gpu','Inches','Product'])
data.head()
```

```python
#data converting to numerical
data = pd.get_dummies(data, dtype=int)
data.head()
```
